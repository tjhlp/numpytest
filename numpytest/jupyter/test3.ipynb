```python
import matplotlib.pyplot as plt
import random
```

```python
x = range(30)
```

```python
y = [random.uniform(25,35) for i in x]
sz = [random.uniform(10, 15) for i in x]
```

```python
plt.figure(figsize=(10,10),dpi=100)
```

```python
plt.plot(x, y,label="北京")
y_ticks = range(40)
plt.plot(x, sz, label="深圳")
x_ticks = ["11点{}分".format(i) for i in x]
plt.yticks(y_ticks[::5])
plt.xticks(x[::5], x_ticks[::5])
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel("时间")
plt.ylabel("温度")
plt.legend(loc="best")
plt.title("中午11点0分到12点之间的温度变化图示", fontsize=15)
```
